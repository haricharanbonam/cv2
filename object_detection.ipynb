```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
newimage = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(newimage)
```

```python
"""
1.splitting image channels
"""
r,g,b=cv2.split(newimage)
print('r : ',r)
print('g : ',g)
print('b : ',b)
```

```python
print('r : ',r.shape)
print('g : ',g.shape)
print('b : ',b.shape)
"""we dont have channels anymore cuz all of them are splitted """
```

```python
"""to merge them we use merge() from cv2"""
new_image = cv2.merge((r,g,b))
```

```python
plt.imshow(new_image)
```

```python
"""Resizing an image"""
s=10
"""s : scaling factor"""
h=int(new_image.shape[0]*s/100)
w=int(new_image.shape[1]*s/100)
"""selects first,second values from new_image.shape (956, 1300, 3)=> w=130,h=95"""
dim=(w,h)
```

```python
resized = cv2.resize(new_image,dim,interpolation=cv2.INTER_AREA)
```

```python
resized.shape
"""(95,130,3)"""
```

```python
plt.imshow(resized)
```

```python
""" Rotating an image """
(h,w,ch)=new_image.shape
center = (w/2,h/2)
"""opencv follows (x,y) so first width and then height format"""
print(center)
angle=90
m=cv2.getRotationMatrix2D(center,angle,1.0)
# Create a 2D rotation matrix centered at 'center' to rotate the image by 'angle' degrees.
"""getRotationMatrix2D(center,angle,scaling factor)"""
rotated=cv2.warpAffine(new_image,m,(w,h))
# Apply the affine rotation matrix to the image and generate the rotated output.
"""warpAffine(image,matrix,dimensions(as we are rotating swap dimen))"""
```

```python
plt.imshow(rotated)
```
